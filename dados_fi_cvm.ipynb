```python
import pandas as pd
import requests
import zipfile

pd.options.display.float_format = '{:.4f}'.format
```

```python
ano = '2022'
mes = '12'
url = f'https://dados.cvm.gov.br/dados/FI/DOC/INF_DIARIO/DADOS/inf_diario_fi_{ano}{mes}.zip'
```

```python
dowload = requests.get(url)

with open(f'inf_diario_fi_{ano}{mes}.zip','wb') as arquivo_cvm:
    
    arquivo_cvm.write(dowload.content)
    
arquivo_zip = zipfile.ZipFile(f'inf_diario_fi_{ano}{mes}.zip')
```

```python
dados_fundos = pd.read_csv(arquivo_zip.open(arquivo_zip.namelist()[0]), sep = ';', encoding = 'ISO-8859-1')
dados_fundos
```

```python
dados_cadastro = pd.read_csv('https://dados.cvm.gov.br/dados/FI/CAD/DADOS/cad_fi.csv', sep=';', encoding='ISO-8859-1')
dados_cadastro = dados_cadastro[['CNPJ_FUNDO', 'DENOM_SOCIAL']]
dados_cadastro = dados_cadastro.drop_duplicates()
dados_cadastro
```

```python
data_inicio_mes = (dados_fundos['DT_COMPTC'].sort_values(ascending = True).unique())[0]
data_fim_mes = (dados_fundos['DT_COMPTC'].sort_values(ascending = True).unique())[-1]

dados_fundos_filtrado = dados_fundos[(dados_fundos['DT_COMPTC'].isin([data_inicio_mes, data_fim_mes]))]

dados_fundos_filtrado
```

```python
base_final = pd.merge(dados_fundos_filtrado, dados_cadastro, how = 'left', left_on = ['CNPJ_FUNDO'], right_on = ['CNPJ_FUNDO'])

base_final = base_final[['CNPJ_FUNDO', 'DENOM_SOCIAL', 'DT_COMPTC', 'VL_QUOTA', 'VL_PATRIM_LIQ', 'NR_COTST']]

base_final
```

```python
fundo_primo_rico = base_final[base_final['DENOM_SOCIAL'].str.contains('ARCA GRÃO', na = False)]

fundo_primo_rico
```

```python
fundo_primo_rico_master = fundo_primo_rico[(fundo_primo_rico['CNPJ_FUNDO'] == '42.847.903/0001-52') & 
(fundo_primo_rico['DT_COMPTC'] == '2022-12-30')]

patrimonio_do_fundo = 'R$ ' + str((fundo_primo_rico_master['VL_PATRIM_LIQ'].iloc[0]/1000000).round(2)) + 'MM'

patrimonio_do_fundo
```

```python
fundo_primo_rico_fic = fundo_primo_rico[(fundo_primo_rico['CNPJ_FUNDO'] == '42.794.402/0001-55')]

retorno_dez = ('Retorno fundo: ' +
              str(((fundo_primo_rico_fic['VL_QUOTA'].iloc[-1]/fundo_primo_rico_fic['VL_QUOTA'].iloc[0] - 1)
                  * 100).round(2))
              + '%')

retorno_dez
```
